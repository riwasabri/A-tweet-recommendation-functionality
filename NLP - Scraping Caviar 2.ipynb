```python
import pickle
import os
import pandas as pd
# Setting variables to be used in format string command below
#tweet_count = 500
text_query = "@Doordash Help"
since_date = "2020-01-01"
until_date = "2022-10-18"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --since {} twitter-search "{} until:{}"> doordash.json'
          .format(since_date, text_query, until_date))
```

```python
df = pd.read_json('doordash.json', lines = True)
```

```python
#retrieving mentioned users to only filter caviar mentions
```

```python
df=df[~df['mentionedUsers'].isna()]
```

```python
def return_username(user):
    list=[]
  
    for i in user:
        list.append(i['username'])
    return list
```

```python
df['mentionedUsers']=df.mentionedUsers.apply(lambda x: return_username(x))
```

```python
def contains_caviar(mentioneduser):
    returned_value='Not set'
    if 'DoorDash_Help' in mentioneduser or 'DoorDash' in mentioneduser:
        if 'Caviar' in mentioneduser:
            returned_value='True'
        else:
            returned_value='False'
    else:
        returned_value='False'
    return returned_value
```

```python
df['Caviar_Mentioned']=df.mentionedUsers.apply(lambda x:contains_caviar(x))
```

```python
df.Caviar_Mentioned.value_counts()
```

```python
df=df[df.Caviar_Mentioned=='True']
```

```python
def contains_influencer(hashtag):
    if '#InfluencerMarketing' in hashtag:
        return 'True'
    else:
        return 'False'
df['influencer']=df.content.apply(lambda x:contains_influencer(x))
```

```python
df=df[(df.influencer=='False')&((df['inReplyToTweetId'].isna()))]
```

```python
df.drop(columns=['Caviar_Mentioned','influencer'],inplace=True)
```
